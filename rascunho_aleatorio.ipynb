```julia
using Distributions
using StatsPlots
```

```julia
Bern = Bernoulli(1/2)
Beta_1_1 = Beta(1, 1)
Beta_5_5 = Beta(5, 5)
Beta_01_05 = Beta(0.1, 0.5)
Beta_1_5 = Beta(1, 5)
Beta_05_01 = Beta(0.5, 0.1)
Beta_5_1 = Beta(5, 1)
Beta_01_01 = Beta(0.1, 0.1)
```

```julia
bern_10 = rand(Bern, 10)
bern_1000 = rand(Bern, 1000)
```

```julia
n = 10
n_ = 1000
soma = sum(bern_10)
soma_ = sum(bern_1000)
```

```julia
Beta_5_5p = Beta(5 + soma, 5 + n - soma)
Beta_5_5p_ = Beta(5 + soma_, 5 + n_ - soma_)
plot(Beta_5_5)
plot!(Beta_5_5p)
plot!(Beta_5_5p_)
```

```julia
Beta_01_05p = Beta(0.1 + soma, 0.5 + n - soma)
Beta_01_05p_ = Beta(0.1 + soma_, 0.5 + n_ - soma_)
plot(Beta_01_05)
plot!(Beta_01_05p)
plot!(Beta_01_05p_)
```

```julia
Beta_1_5p = Beta(1 + soma, 5 + n - soma)
Beta_1_5p_ = Beta(1 + soma_, 5 + n_ - soma_)
plot(Beta_1_5)
plot!(Beta_1_5p)
plot!(Beta_1_5p_)
```

```julia
Beta_05_01p = Beta(0.5 + soma, 0.1 + n - soma)
Beta_05_01p_ = Beta(0.5 + soma_, 0.1 + n_ - soma_)
plot(Beta_05_01)
plot!(Beta_05_01p)
plot!(Beta_05_01p_)
```

```julia
Beta_5_1p = Beta(5 + soma, 1 + n - soma)
Beta_5_1p_ = Beta(5 + soma_, 1 + n_ - soma_)
plot(Beta_5_1)
plot!(Beta_5_1p)
plot!(Beta_5_1p_)
```

```julia
Beta_01_01p = Beta(0.1 + soma, 0.1 + n - soma)
Beta_01_01p_ = Beta(0.1 + soma_, 0.1 + n_ - soma_)
plot(Beta_01_01)
plot!(Beta_01_01p)
plot!(Beta_01_01p_)
```

```julia
Beta_1_1p = Beta(1 + soma, 1 + n - soma)
Beta_1_1p_ = Beta(1 + soma_, 1 + n_ - soma_)
plot(Beta_1_1)
plot!(Beta_1_1p)
plot!(Beta_1_1p_)
```

```julia
dists = [Beta_5_5, Beta_5_5p, Beta_5_5p_, Beta_01_05, Beta_01_05p, Beta_01_05p_, Beta_1_5, Beta_1_5p, Beta_1_5p_, Beta_05_01, Beta_05_01p, Beta_05_01p_, Beta_5_1, Beta_5_1p, Beta_5_1p_, Beta_01_01, Beta_01_01p, Beta_01_01p_, Beta_1_1, Beta_1_1p, Beta_1_1p_]
```

```julia
medias = [mean(i) for i in dists]
```

```julia
variancia = [var(i) for i in dists]
```

```julia
modas = zeros(21)
modas = Array{Any}(modas)
for (i, j) in enumerate(dists) 
    try 
        modas[i] = mode(j) 
    catch
        modas[i] = missing 
    end
end 
```
