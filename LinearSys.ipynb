```julia
function triang_matrix(n)
    A = rand(Float32, n, n)
    @simd for i in 1:(n-1)
        @inbounds A[i, (i+1):n] .= 0
    end
    return A
end
```

```julia
function sys_linear(A, B)
    X = zeros(length(B))
    X[1] = B[1]/A[1,1]
    for i in 2:length(B)
        soma = 0
        @inbounds el = A[i, 1:(i-1)] .* X[1:(i-1)]
        soma += sum(el)
        @inbounds X[i] = (B[i] - soma)/A[i,i]
    end
    return X
end
```

```julia
using BenchmarkTools
```

```julia
function teste(n)
    A = triang_matrix(n)
    B = rand(n)
    return sys_linear(A, B)
end
```

```julia
teste1 = @benchmark teste(8)
```

```julia
teste2 = @benchmark teste(16) 
```
