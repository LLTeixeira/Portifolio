```julia
using StatsPlots
```

```julia
n1 = 10
s1 = sum(amostra_10)
n2 = 1000
s2 = sum(amostra_1000)
```

```julia
D1 = Gamma(1, 1)
plot(D1, xlim= (0, 2))
D1p1 = Gamma(1 + n1, (1 + s1)^-1)
plot!(D1p1)
D1p2 = Gamma(1 + n2, (1 + s2)^-1)
plot!(D1p2)
```

```julia
alpha = 10
beta = 10
D2 = Gamma(alpha, beta^-1)
plot(D2, xlim= (0, 2))
D2p1 = Gamma(alpha + n1, (beta + s1)^-1)
plot!(D2p1)
D2p2 = Gamma(alpha + n2, (beta + s2)^-1)
plot!(D2p2)
```

```julia
alpha = 5
beta = 5
D3 = Gamma(alpha, beta^-1)
plot(D3, xlim= (0, 2))
D3p1 = Gamma(alpha + n1, (beta + s1)^-1)
plot!(D3p1)
D3p2 = Gamma(alpha + n2, (beta + s2)^-1)
plot!(D3p2)
```

```julia
alpha = 5
beta = 1
D4 = Gamma(alpha, beta^-1)
plot(D4, xlim= (0, 6))
D4p1 = Gamma(alpha + n1, (beta + s1)^-1)
plot!(D3p1,)
D4p2 = Gamma(alpha + n2, (beta + s2)^-1)
plot!(D4p2)
```

```julia
alpha = 5
beta = 10
D5 = Gamma(alpha, beta^-1)
plot(D5, xlim= (0, 1.5))
D5p1 = Gamma(alpha + n1, (beta + s1)^-1)
plot!(D3p1,)
D5p2 = Gamma(alpha + n2, (beta + s2)^-1)
plot!(D5p2)
```

```julia
alpha = 10
beta = 20
D6 = Gamma(alpha, beta^-1)
plot(D6, xlim= (0, 1.5))
D6p1 = Gamma(alpha + n1, (beta + s1)^-1)
plot!(D3p1,)
D6p2 = Gamma(alpha + n2, (beta + s2)^-1)
plot!(D6p2)
```

```julia
Distribuicoes = [D1, D1p1, D1p2, D2, D2p1, D2p2, D3, D3p1, D3p2, D4, D4p1, D4p2, D5, D5p1, D5p2, D6, D6p1, D6p2]
```

```julia
Alphas = [shape(i) for i in Distribuicoes]
Betas = [(scale(i))^-1 for i in Distribuicoes]
Medias = [mean(i) for i in Distribuicoes]
Variancias = [var(i) for i in Distribuicoes]
```
