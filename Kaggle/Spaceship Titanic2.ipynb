```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import RobustScaler
from sklearn.impute import KNNImputer
```

```python
def preprocessamento(train, test):
    data = [train, test]
    encoder = OneHotEncoder(handle_unknown='ignore', drop='if_binary', sparse_output=False)
    scaler = RobustScaler()
    imputer = KNNImputer(n_neighbors=5)
    for j, k in enumerate(data):
        numericas = k[["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]]
        categoricas = k[["Name", "HomePlanet", "CryoSleep", "Destination", "VIP"]]
        cabines = k.Cabin
        lista = []
        for i in cabines:
            if type(i) == str:
                lista.append(i.split("/"))
            else:
                lista.append([i, i, i])
        lista = np.array(lista)
        categoricas["Cabin_deck"] = lista[:, 0]
        numericas["Cabin_room"] = lista[:, 1]
        categoricas["Cabin_side"] = lista[:, 2]
        
        if j == 0:
            categoricas = encoder.fit_transform(categoricas)
            numericas = scaler.fit_transform(numericas)
            dados = np.hstack((numericas, categoricas))
            train = imputer.fit_transform(dados)
        elif j == 1:
            categoricas = encoder.transform(categoricas)
            numericas = scaler.transform(numericas)
            dados = np.hstack((numericas, categoricas))
            test = imputer.transform(dados)
    return [train, test]
```

```python
dados = preprocessamento(train_data, test_data)
```

```python
test = dados[1]
test.shape
```

```python
train = dados[0]
train.shape
```

```python
log_reg = LogisticRegression(random_state=2023, solver='lbfgs', max_iter=100000).fit(X_train, y_train)
log_reg.score(X_val, y_val)
```

```python
adab = AdaBoostClassifier(n_estimators=650, random_state=10).fit(X_train, y_train)
adab.score(X_val, y_val)
```

```python
gbc7 = GradientBoostingClassifier(n_estimators=1320, random_state=10).fit(X_train, y_train)
gbc7.score(X_val, y_val)
```
