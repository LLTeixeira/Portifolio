```python
from sklearn.preprocessing import OneHotEncoder
#from sklearn.preprocessing import QuantileTransformer
from sklearn.impute import KNNImputer
```

```python
def preprocessamento(train, test):
    data = [train, test]
    encoder = OneHotEncoder(handle_unknown='ignore', drop='if_binary', sparse_output=False)
    #qt = QuantileTransformer(n_quantiles=4277, random_state=2023)
    imputer = KNNImputer(n_neighbors=5)
    for j, k in enumerate(data):
        numericas = k[["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]]
        categoricas = k[["Name", "HomePlanet", "CryoSleep", "Destination", "VIP"]]
        cabines = k.Cabin
        lista = []
        for i in cabines:
            if type(i) == str:
                lista.append(i.split("/"))
            else:
                lista.append([i, i, i])
        lista = np.array(lista)
        categoricas["Cabin_deck"] = lista[:, 0]
        numericas["Cabin_room"] = lista[:, 1]
        categoricas["Cabin_side"] = lista[:, 2]
        
        if j == 0:
            categoricas = encoder.fit_transform(categoricas)
            #numericas = qt.fit_transform(numericas)
            dados = np.hstack((numericas, categoricas))
            train = imputer.fit_transform(dados)
        elif j == 1:
            categoricas = encoder.transform(categoricas)
            #numericas = qt.transform(numericas)
            dados = np.hstack((numericas, categoricas))
            test = imputer.transform(dados)
    return [train, test]
```

```python
dados = preprocessamento(train_data, test_data)
```

```python
import tensorflow as tf
input_size = X_train.shape[1]
output_size = 1
```

```python
inputs = tf.keras.Input(shape=(input_size,))
a1 = tf.keras.layers.Dense(1024, activation='relu')(inputs)
a2 = tf.keras.layers.Dense(2048, activation='relu')(a1)
a3 = tf.keras.layers.Dense(2048, activation='relu', kernel_regularizer='l2')(a2)
a4 = tf.keras.layers.Dense(1024, activation='relu', kernel_regularizer='l2')(a3)
a5 = tf.keras.layers.Dense(512, activation='relu')(a4)
a6 = tf.keras.layers.Dense(256, activation='relu')(a5)
a7 = tf.keras.layers.Dense(128, activation='relu')(a6)
#dropout = tf.keras.layers.Dropout(rate=0.15)(a7)
outputs = tf.keras.layers.Dense(output_size, activation='sigmoid')(a7)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=5e-4),
              loss = tf.keras.losses.BinaryCrossentropy(),
              metrics=[tf.keras.metrics.BinaryAccuracy()])

#tf.keras.utils.plot_model(model, show_shapes=True, dpi=80)      
```

```python
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=128,
    epochs=100,
    verbose=1,
    validation_data=(X_val, y_val),
    shuffle=True,
    callbacks=False
)
```
