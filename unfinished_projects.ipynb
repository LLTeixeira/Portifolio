```julia
using Random
using StatsBase
```

```julia
function ball()
    position = 0 
    for _ in 1:1000
        roll = sample([-1, 1], Weights([0.5, 0.5]), 1)[1]
        position += roll
    end
    return position
end
```

```julia
function simulation()
    data = zeros(Int32, 100000)
    for i in 1:length(data)
        data[i] = ball()
    end
    return data
end
```

```julia
data = simulation()
```

```julia
using Plots
```

```julia
gr()
histogram(data)
```

```julia
function ball_skewed()
    position = 0 
    for _ in 1:1000
        roll = sample([-1, 1], Weights([0.3, 0.7]), 1)[1]
        position += roll
    end
    return position
end

function simulation_skewed()
    data = zeros(Int32, 100000)
    for i in 1:length(data)
        data[i] = ball_skewed()
    end
    return data
end
```

```julia
data_skewed = simulation_skewed()
```

```julia
histogram(data_skewed)
```

```julia
mean_and_std(data)
```

```julia
kurtosis(data)
```

```julia
mean_and_std(data_skewed)
```

```julia
kurtosis(data_skewed)
```
