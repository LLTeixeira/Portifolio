```julia
using Pingouin
```

```julia
function ball(wheight)
    position = 0 
    for _ in 1:1000
        roll = sample([-1, 1], Weights([wheight, 1-wheight]), 1)[1]
        position += roll
    end
    return position
end
```

```julia
function simulation(wheight)
    data = zeros(Int32, 100000)
    for i in 1:length(data)
        data[i] = ball(wheight)
    end
    return data
end
```

```julia
data = simulation(0.5)
```

```julia
gr()
histogram(data)
```

```julia
data_skewed = simulation(0.2)
```

```julia
histogram(data_skewed)
```

```julia
mean_and_std(data)
```

```julia
kurtosis(data)
```

```julia
mean_and_std(data_skewed)
```

```julia
kurtosis(data_skewed)
```

```julia
function mean_variation()
    medias = []
    for i in 0.1:0.1:0.9
        media = mean(simulation(i))
        append!(medias, media)
    end
    return medias
end
```

```julia
medias = mean_variation()
```

```julia
x = collect(0.1:0.1:0.9)
scatter(x, medias)
```

```julia
new_point = mean(simulation(0.35))
```

```julia
append!(medias, new_point)
append!(x, 0.35)
```

```julia
scatter(x, medias)
```

```julia
function ball_alternating()
    position = 0 
    for i in 1:1000
        roll = sample([-1, 1], Weights([abs(sin(i)), 1-abs(sin(i))]), 1)[1]
        position += roll
    end
    return position
end
```

```julia
function simulation_alternating()
    data = zeros(Int32, 100000)
    for i in 1:length(data)
        data[i] = ball_alternating()
    end
    return data
end
```

```julia
alternating = simulation_alternating()
```

```julia
gr()
histogram(alternating)
```

```julia
mean_and_std(alternating)
```

```julia
function ball_alternating_pog()
    position = 0 
    y(x) = (sin(x)+1)/2
    for i in 1:1000
        roll = sample([-1, 1], Weights([y(i), 1-y(i)]), 1)[1]
        position += roll
    end
    return position
end
function simulation_alternating_pog()
    data = zeros(Int32, 100000)
    for i in 1:length(data)
        data[i] = ball_alternating_pog()
    end
    return data
end
```

```julia
pog = simulation_alternating_pog()
```

```julia
histogram(pog)
```

```julia
mean_and_std(pog)
```

```julia
normality(pog, method="jarque_bera")
```

```julia
function ball_exp()
    position = 0 
    y(x) = (ℯ^(-x))
    for i in 1:1000
        i /= 628
        roll = sample([-1, 1], Weights([y(i), 1-y(i)]), 1)[1]
        position += roll
    end
    return position
end
function simulation_exp()
    data = zeros(Int32, 100000)
    for i in 1:length(data)
        data[i] = ball_exp()
    end
    return data
end
```

```julia
exp = simulation_exp()
```

```julia
histogram(exp)
```

```julia
mean_and_std(exp)
```
